```python
"""Test the trained policy (for REINFORCE).

Author: Elie KADOCHE.
"""

import torch
import gymnasium as gym
import numpy as np
from eve_to_students.src.models.actor import ActorModel


# Create policy
policy = ActorModel()
policy.eval()
print(policy)

# Load the trained policy
policy = torch.load("reinforce/models/content.pt", map_location=torch.device('cpu'))

# Create environment
env = gym.make("LunarLander-v2")
env._max_episode_steps = 500

# Reset it
total_reward = 0.0
state, _ = env.reset(seed=None)

play = []
row = {'X Coordinate': 0, 'Y Coordinate': 0, 'Linear Velocity X': 0, 'Linear Velocity Y': 0,
       'Angle': 0, 'Angular Velocity': 0, 'Leg1': 0, 'Leg2': 0, 'Action': 0, 'Reward': 0, 'Episode':0}


# While the episode is not finished

for ep in range(500):
    env.reset()
    terminated = False
    truncated = False
    while not terminated and not truncated:

        # Use the policy to generate the probabilities of each action
        probabilities = policy(state)

        # ---> TODO: how to select an action
        # select action with the greatest probability according to policy
        action = int(torch.argmax(probabilities))

        # One step forward
        state, reward, terminated, truncated, _ = env.step(action)
        
        rowlist = np.concatenate((state, [action], [reward]))
        row.update(zip(row, rowlist))
        row['Episode'] = ep
        play.append(row.copy())
```
