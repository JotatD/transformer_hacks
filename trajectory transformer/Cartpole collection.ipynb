```python
import gymnasium as gym
import pandas as pd
import numpy as np
env = gym.make('CartPole-v1')



data_collection = []

for i in range(1000):
    observation = env.reset()
    done = False
    time = 0
    while not done and time <= 500:
        action = env.action_space.sample()
        next_observation, reward, done, info, _ = env.step(action)
        reward = np.array(reward).reshape(1)
        episode = np.array(i).reshape(1)
        action = np.array(action).reshape(1)
        epi = np.concatenate((episode, next_observation, action,reward))
        
        data_collection.append(epi)
        
        
        if done:
            break
        time += 1

env.close()
```

```python
from src.models.actor import ActorModel
import torch
policy = ActorModel()
policy.eval()
print(policy)

# Load the trained policy
policy = torch.load("models/actor_bis.pt")
env = gym.make('CartPole-v1')
# Create environment


for i in range(1000, 1020):
    total_reward = 0
    state, _ = env.reset(seed=None)
    terminated = False

    while not terminated:
        # Use the policy to generate the probabilities of each action
        probabilities = policy(state)
        # select action with the greatest probability according to policy
        action = int(torch.argmax(probabilities))

        # One step forward
        state, reward, terminated, done, info = env.step(action)

        reward = np.array(reward).reshape(1)
        episode = np.array(i).reshape(1)
        action = np.array(action).reshape(1)
        epi = np.concatenate((episode, next_observation, action, reward))

        data_collection.append(epi)
        time += 1
        # Render (or not) the environment
        total_reward += reward
        if done:
            break

    # Print reward
    print("total_reward = {}".format(total_reward))
```
